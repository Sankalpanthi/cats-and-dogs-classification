# Predicting using test images

```python
import cv2
import keras
import matplotlib.pyplot as plt
```

```python
import numpy as np
CATEGORIES = ['Cat', 'Dog']
```

## Processing the test image

```python
def image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is not None:
        #plt.imshow(img)
        new_arr = cv2.resize(img, (60, 60))
        new_arr = np.array(new_arr)
        new_arr = new_arr.reshape(-1, 60, 60, 1)
        return new_arr
    else:
        return None
```

```python
model = keras.models.load_model('3x3x64-catvsdog.model')
```

```python
import os
path = r'C:\Users\user\Desktop\cats vs dogs\dogscats\test2' #raw string
output = []
```

```python
for img in os.listdir(path):
        img_path = os.path.join(path, img)
        test=image(img_path)
        if test is not None:
            prediction = model.predict(test) # getting result using model
            output.append([img,CATEGORIES[prediction.argmax()]])
        else:
            print("Error")
```

```python
for i,o in output:
    print("Image ",i[:-4],": ",o)
```
